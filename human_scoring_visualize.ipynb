```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming scores_with_variance and models are already defined
models = list(scores_with_variance.keys())  # Model names
total_std_devs = []

# Define a list of colors for the bars, one for each model
colors = ['blue', 'orange', 'green', 'red', 'orange', 'cyan']

# Make sure we have enough colors for the models
assert len(colors) >= len(models), "Define more colors to match the number of models."

# Calculate total standard deviation for each model
for model, data in scores_with_variance.items():
    total_std = sum(data['std_devs'])  # Sum of standard deviations across all categories for the model
    total_std_devs.append(total_std)

# X locations for the models
ind = np.arange(len(models))

fig, ax = plt.subplots(figsize=(4, 6))

# Plot total standard deviations for each model, each with its own color
for i, (model, total_std) in enumerate(zip(models, total_std_devs)):
    ax.bar(ind[i], total_std, width=0.35, label=model, color=colors[i])

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Models')
ax.set_ylabel('Total Standard Deviation')
ax.set_title('Total Standard Deviations Across All Categories for Each Model')
ax.set_xticks(ind)
ax.set_xticklabels(models, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
